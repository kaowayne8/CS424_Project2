```r
#Read in CSV files
rider <- read.csv(file="CTA_Ridership_CSV.csv", header=TRUE, stringsAsFactors = FALSE)
info <- read.csv(file="CTA_Info_CSV.csv", header=TRUE, stringsAsFactors = FALSE)
```

```r
merged_data <- merge(rider, info, by.x="station_id", by.y="MAP_ID", all.x= TRUE)
```

```r
merged_data
```

```r
typeof(merged_data$Location)
```

```r
#Stations that don't have coordinate (Just test to make sure data parced correctly)
unique(merged_data[is.na(merged_data$STOP_NAME),]$stationname)
merged_data[merged_data$stationname == "Homan" & merged_data$rides != 0,]

RaWa <- "(41.884431, 87.626149)"
WaSt <- '(41.8837, 87.6278)'
MaWa <- "(41.882023, 87.626098)"
Ho <- "(41.884914, 87.711327)"
#Fix coordinates of Randolph/Wabash
merged_data$Location[merged_data$stationname == "Randolph/Wabash"] <- RaWa
#Fix coordinates of Washington/State
merged_data$Location[merged_data$stationname == "Washington/State"] <- WaSt
#Fix coordinates of Madison/Wabash
merged_data$Location[merged_data$stationname == "Madison/Wabash"] <- MaWa
#Fix coordinates of Homan
merged_data$Location[merged_data$stationname == "Homan"] <- Ho

# At this point, merged_data shouldn't have any NA's in Location
```

```r
simple <- data.frame(merged_data$station_id, merged_data$stationname, merged_data$date, merged_data$daytype, merged_data$rides, merged_data$STOP_ID, merged_data$Location)
colnames(simple) <- c("station_id", "stationname", "date", "daytype", "rides", "STOP_ID", "Location")
#  simple #Just returns the minimum amount of columns needed. Can change later if needed
simple$date <- mdy(simple$date)
simple
```
